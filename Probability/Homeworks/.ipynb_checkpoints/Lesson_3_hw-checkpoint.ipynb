### Задание 1

**Из 500 компьютеров  180 принадлежат к 1 партии, 170 – ко второй партии, остальные к третьей. В первой партии 3% брака, во второй — 2%, в третьей – 6%. Случайно выбирается один компьютер. 500 - 180 - 170 = 150 третья партия

3% от 180 = 5.4 брак по первой партии

2% от 170 = 3.4 брак по второй партии

6% от 150 = 9 брак по третьей партии

17.8 / 500 = 0.0356 вероятность случайно выбрать бракованный компьютер

### Задание 4

**Есть 4 кубика. На трех из них окрашена белым половина граней, а на четвертом кубике всего одна грань из шести белая. Наудачу выбранный кубик подбрасывается семь раз. Найти вероятность того, что был выбран четвертый кубик, если при семи подбрасываниях белая грань выпала ровно один раз.  Ответ округлите до тысячных. Используйте точку в качестве десятичного разделителя.**

```python
from math import factorial
```

```python
# кубик с одной белой гранью
def cube_1(n, k):
    p, q = 1/6, 5/6
    P = factorial(n)/(factorial(k) * factorial(n - k))
    return P * (p ** k) * (q ** (n - k))

cube_1(n=7, k=1)
```

```python
# кубик с 3 белыми и 3 черными гранями
def cube_2(n, k):
    p, q = 1/2, 1/2
    P = factorial(n)/(factorial(k) * factorial(n - k))
    return P * (p ** k) * (q ** (n - k))

cube_2(n=7, k=1)
```

```python
1/4 * cube_1(n=7, k=1) + 3/4 * cube_2(n=7, k=1)
```

```python
(1/4 * cube_1(n=7, k=1)) / 0.1386942015317787
```

Ответ: 0.704
