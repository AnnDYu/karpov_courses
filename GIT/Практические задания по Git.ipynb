# Домашние задания по GIT.

**1. Что делает команда cd?**

* Выводит "Hello, World!"
* Копирует файлы!
* **Перемещает по папкам, это очевидно!**
* Открывает дисковод компьютера

**2. Для чего нужна команда cat?**

* Открыть браузер с запросом котики
* Сконкатенировать датафрэймы в один
* **Вывести содержимое файла или файлов**
* Открыть дисковод!

**3. А команда pwd?**

* Ну теперь-то точно для открывания дисковода
* **Вывести путь папки, где вы находитесь**
* Сокращение от purge working directory - очистит её от незакомиченных файлов
* Переместить файлы

**4. Перейдём к более серьёзным вещам – что делает команда rm?**

* Печатает директорию, где мы находимся
* Создаёт пустую директорию
* **Необратимо удаляет файлы**
* Запускает терминал

**5. Что делает git clone?**

* Создаёт копию git'а на компьютере
* Перенесёт репозиторий из гитлаба на компьютер
* Копирует описание репозитория
* **Правильно будет git clone some_link — эта команда создаст копию репозитория, хранящегося по адресу some_link**

**6. Что делает git status?**

* **Выводит статус репозитория**
* Удаляет файл status
* Скачивает репозиторий на компьютер
* Добавляет файл status в коммит

**7. А что сделает команда git pull?**

* Склонирует репозиторий на машину
* Обновит состояние репозитория до последнего общепринятого состояния
* **Правильно будет git pull some_link — это обновит состояние указанного по адресу some_link репозитория до последнего общепринятого состояния**
* Обновит общепринятое состояние репозитория до нашего последнего изменения

**8. Что делает git branch?**

* Создаёт новую копию репозитория
* Создаёт новую ветвь
* **Выводит список имеющихся веток**
* Коммитит изменения репозитория в новую ветку

**9. А что делает эта команда?**

**git branch my_branch**

* Эта команда даст ошибку!
* Выводит существующие ветви
* Переводит нас в ветку my_branch
* **Создаёт ветвь my_branch**

**10. Для чего нужна команда git add?**

* **Правильно будет git add file или git add . — это добавит изменённые файлы в коммит**
* Правильно будет git add username — это добавит пользователя в список тех, кто может коммитить
* Эта команда добавляет файлы в коммит
* Команда добавляет изменённые файлы в главный репозиторий на гитлабе

**11. Начнем с простого: вы пришли на работу, и, чтобы сделать очень важную задачу, вам потребуется новая ветка, где вы будете писать код.**

Вы хотите:

1. перейти в ветку master
2. стянуть все самые последние обновления 
3. создать новую ветку my_new_branch
4. перейти в ветку my_new_branch

Как будет выглядеть правильная последовательность команд?

```
1. git checkout master
2. git pull
3. git branch my_new_branch
4. git checkout my_new_branch
```

**12. Теперь вы:**

**1. написали новый скрипт my_script.py в ветке my_new_branch и закоммитили результаты с сообщением 'обновил логику считывания данных'**

**2. Однако перед тем, как создать merge request, вы решили обновить вашу ветку my_new_branch последними обновлениями из ветки master**

**3. Для этого вам нужно перейти в ветку master, обновить ветку master, а потом вмёрджить ветку master в ветку my_new_branch**

**4. После этого вы создали merge request**

**Укажите, какая последовательность команд даст нужный результат:**

1. git add my_script.py
2. git commit -m 'обновил логику считывания данных'
3. git checkout master
4. git pull
5. git checkout my_new_branch
6. git merge master
7. git push origin my_new_branch
